# 04 — Side-by-side Comparison

Same corpus, same questions, three pipelines. This is the educational payoff — a single table where you can read off where each pattern wins and loses.

```python
from rag_compare.common import load_corpus
from rag_compare.kg_rag import KGRagPipeline
from rag_compare.hybrid_rag import HybridRagPipeline
from rag_compare.agentic_rag import AgenticRagPipeline
from rag_compare.eval import GOLDEN_QUESTIONS

docs = load_corpus()
kg = KGRagPipeline.build(docs)
hybrid = HybridRagPipeline.build(docs)
agent = AgenticRagPipeline.build(hybrid=hybrid, kg=kg)
```

```python
import pandas as pd

rows = []
for item in GOLDEN_QUESTIONS:
    rows.append({
        "capability": item.capability,
        "question": item.question,
        "expected": item.expected,
        "kg_rag": kg.query(item.question),
        "hybrid_rag": hybrid.query(item.question),
        "agentic_rag": agent.query(item.question),
    })

df = pd.DataFrame(rows)
df
```

## What to look for

- **Multi-hop** rows: KG-RAG and Agentic should outperform Hybrid.
- **Exact-term** rows: Hybrid should do well; pure dense baselines fail here.
- **Unanswerable** rows: which pipelines hallucinate? Faithfulness is checked properly in notebook 05.
